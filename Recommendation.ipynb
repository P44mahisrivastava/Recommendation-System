```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/IMDB_Top250Engmovies2_OMDB_Detailed.csv")
df.head(2)
```

```python
new_df=df[['Title','Genre','Director','Actors','Plot']]
new_df.head()
```

```python
pip install rake-nltk==1.0.0
```

```python
from rake_nltk import Rake
```

```python
new_df['Key_words']=""
```

```python
for index,row in new_df.iterrows():
  plot=row['Plot']
  r=Rake()
  r.extract_keywords_from_text(plot)
  key_words_dict_scores=r.get_word_degrees()
  row['Key_words']=list(key_words_dict_scores.keys())
```

```python
new_df=new_df.drop(['Plot'],axis=1)

```

```python
new_df
```

```python
new_df['Actors']=new_df['Actors'].map(lambda x:x.split(',')[:3])
```

```python
new_df['Genre']=new_df['Genre'].map(lambda x:x.split(','))
```

```python
new_df['Director']=new_df['Director'].map(lambda x:x.lower().split(' '))
```

```python
for index,row in new_df.iterrows():
    row['Actors']=[x.lower().replace(' ','') for x in row['Actors']]
    row['Director']=''.join(row['Director'])

```

```python
new_df
```

```python
new_df['Director']=new_df['Director'].apply(lambda x:x.split()) #converting string to list
new_df['tags']=new_df['Genre']+new_df['Director']+new_df['Actors']+new_df['Key_words'] #merging all columns
new_df
```

```python
new_df=new_df.drop(columns=['Genre','Director','Key_words','Actors'])
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count=CountVectorizer()
count_matrix=count.fit_transform(new_df['tags']).toarray()
count.get_feature_names()
len(count.get_feature_names())
```

```python
count_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim=cosine_similarity(count_matrix)
cosine_sim[0]
```
